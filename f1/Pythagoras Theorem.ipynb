```python
from manim import *
```

```python
%%manim -qm MovingFrameBox

class MovingFrameBox(Scene):
    def construct(self):
        
        display = Text("Verfiying Pythagoras' Theorem", gradient=(RED, BLUE, GREEN), font_size=70)
        self.play(Create(display))
        self.play(FadeOut(display))
        
        base = Line([-1,-2,0], [2,-2,0]).set_color(ORANGE)
        perp = Line([-1,-2,0], [-1,2,0]).set_color(BLUE) 
        hyp  = Line([2,-2,0], [-1,2,0]).set_color(TEAL)
    
        self.play(Create(base))
        self.play(Create(perp))
        self.play(Create(hyp)) 
        
        basetext = Text('3 unit').next_to(base, DOWN)
        perptext = Text('4 unit').next_to(perp, LEFT)
        hyptext  = Text('5 unit').next_to(perp, 10*RIGHT)
        
        self.play(Create(basetext))
        self.play(Create(perptext))
        self.play(Create(hyptext))
        self.wait(2)
        self.play(FadeOut(perptext,basetext, hyptext))
        
        
        self.play(perp.animate.shift(2*LEFT))
        self.wait(2)
        square1= Square(4,grid_xstep=1.0, grid_ystep= 1.0).next_to(perp,LEFT).set_fill(BLUE, opacity=0.5).scale(0.75)
        self.play(Transform(perp,square1))
        
        
        self.play(hyp.animate.shift(RIGHT))
        self.wait(2)
        square2= Square(5,grid_xstep=1.0, grid_ystep= 1.0).next_to(hyp,RIGHT).set_fill(TEAL, opacity=0.5).scale(0.75).shift(LEFT)
        self.play(Transform(hyp,square2))
        
        
        
        self.play(base.animate.shift(UP))
        square3= Square(3,grid_xstep=1.0, grid_ystep= 1.0).next_to(base,UP).set_fill(ORANGE, opacity=0.5).scale(0.75).shift(DOWN+0.75*LEFT)
        self.play(Transform(base,square3))
        
        b1 = Brace(square1)
        b2 = Brace(square3)
        b3 = Brace(square2)
        
        self.play(Create(b1))
        b1text = Text('16 sq. unit').next_to(b1, DOWN)
        self.play(Create(b1text))
        
        
        plus = Text('+').next_to(square1, 3*RIGHT).scale(2)
        self.play(Create(plus))
        
        self.play(Create(b2))
        b2text = Text('9 sq. unit ').next_to(b2, DOWN)
        self.play(Create(b2text))
        
        equals = Text('=').next_to(square3,2*RIGHT).scale(2)
        self.play(Create(equals))
        
        self.play(Create(b3))
        b3text = Text('25 sq. unit').next_to(b3, DOWN)
        self.play(Create(b3text))
```
